# Exporting data to mySql Server 

```python
from sqlalchemy import create_engine
import pymysql
# MySQL connection details
username = 'root'
password = '1234'
host = 'localhost'  # or your MySQL server's IP
port = '3306'       # Default MySQL port
database = 'Census_2011'  # Replace with your database name

# Create SQLAlchemy engine
connection_string = f'mysql+pymysql://{username}:{password}@{host}:{port}/{database}'
engine = create_engine(connection_string)

# Write DataFrame to MySQL
df.to_sql(
    name='census',  # Replace with the desired table name
    con=engine,
    if_exists='append',     # Replace the table if it already exists
    index=False              # Avoid writing the DataFrame index as a column
)
# Close the connection
engine.dispose()
```

# Finding state with highest popultion

```python
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/Census_2011")
querry="""
SELECT State_name, SUM(Population) as Population
from census 
GROUP BY State_name
ORDER BY Population;
"""
df=pd.read_sql(querry,engine)
engine.dispose()

plt.figure(figsize=(10,8))
sns.barplot(data=df,x="State_name",y="Population",hue="State_name",palette="viridis")
plt.ylabel("Population",fontsize=10,weight="bold")
plt.xlabel("Sates",fontsize=10,weight="bold")
plt.xticks(rotation=90)
plt.title("State wise Population", fontsize=14,weight="bold")
plt.show()
```

# Gender ratio


```python
# Modify the SQL query to select both Male and Female population and calculate Gender Ratio
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/Census_2011")
query = """
select State_name, sum(Male) as Male, sum(Female) as Female,
       sum(Female) / sum(Male) * 1000 as Gender_ratio
from census
group by State_name
order by State_name;
"""

# Execute the query and load the data into a dataframe
df = pd.read_sql(query, engine)

# Close the engine connection
engine.dispose()

# Reshape the data into a long format suitable for seaborn
df_melted = df.melt(id_vars=["State_name", "Gender_ratio"], value_vars=["Male", "Female"], 
                    var_name="Gender", value_name="Population")

# Plotting the gender ratio with clustered bars for Male and Female populations
plt.figure(figsize=(14, 8))

# Create the clustered bar chart
sns.barplot(x='State_name', y='Gender_ratio', hue='Gender', data=df_melted)

# Set chart labels and title
plt.xlabel('State Name', fontsize=12, weight='bold')
plt.ylabel('Gender Ratio (Females per 1000 Males)', fontsize=12, weight='bold')
plt.title('Gender Ratio (Females per 1000 Males) Across States', fontsize=14, weight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

# Calculating State-Wise Literacy Rate 

```python
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/Census_2011")
querry="""
select State_name, sum(Literate) / sum(Population) * 100 as Literacy_Rate
from census
group by State_name
order by Literacy_Rate desc;
"""
df=pd.read_sql(querry,engine)
engine.dispose()

plt.figure(figsize=(10,8))
sns.barplot(data=df,x="State_name", y="Literacy_Rate")
plt.xlabel("State Name",fontsize=12,weight="bold")
plt.ylabel("literacy Rate in %", fontsize=12,weight="bold")
plt.title("State Wise Literacy Rate",fontsize=14,weight="bold")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Calculating Secondary Education Rate And Higher Education Rate 

```python
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/Census_2011")
querry="""
select State_name,
		(sum(Secondary_Education) / sum(Population) )* 100 as Secondary_Education_Rate,
        (sum(Higher_Education) / sum(Population)) * 100 as Higher_Education_Rate
from census
group by State_name
order by Secondary_Education_Rate desc, Higher_Education_Rate desc;
"""

df=pd.read_sql(querry,engine)
engine.dispose()

df_melted = df.melt(id_vars=["State_name"], 
                    value_vars=["Secondary_Education_Rate", "Higher_Education_Rate"],
                    var_name="Education_Level", value_name="Education_Rate")

# Plotting the data
plt.figure(figsize=(14, 8))

sns.barplot(x='State_name', y='Education_Rate', hue='Education_Level', data=df_melted)

# Adding labels and title
plt.xlabel('State Name', fontsize=12, weight='bold')
plt.ylabel('Education Rate (%)', fontsize=12, weight='bold')
plt.title('Secondary and Higher Education Rates Across States', fontsize=14, weight='bold')

# Rotating x-axis labels for better readability
plt.xticks(rotation=90, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

# Finding States with Working Women Percentage 

```python
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/Census_2011")
querry="""
select State_name, 
	(sum(Female_Workers) / sum(Female)) * 100 as Working_Women_Rate
from census
group by State_name
order by Working_Women_Rate desc;
"""
df=pd.read_sql(querry,engine)
engine.dispose()

plt.figure(figsize=(10,8))
sns.barplot(data=df,x="State_name",y="Working_Women_Rate")
plt.xlabel("State name",fontsize=12,weight="bold")
plt.ylabel("Working Women Percentage",fontsize=12,weight="bold")
plt.title("State Wise Working Women Percentage",fontsize=14,weight="bold")
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

# Finding Trends in Cultivators v/s Agricultural Workers 

```python
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/Census_2011")
querry="""
select State_name,
	sum(Cultivator_Workers) as Cultivators,
    sum(Agricultural_Workers) as Agricultural_Workers
from census
group by State_name
order by State_name;
"""
df=pd.read_sql(querry,engine)
engine.dispose()

df_melted = df.melt(id_vars=["State_name"], 
                    value_vars=["Cultivators", "Agricultural_Workers"],
                    var_name="Worker_Type", value_name="Number_of_Workers")

plt.figure(figsize=(10,8))
sns.barplot(data=df_melted,x="State_name", y="Number_of_Workers",hue="Worker_Type")
plt.xlabel("State name",fontsize=12,weight="bold")
plt.ylabel("Number of Workers",fontsize=12,weight="bold")
plt.title("Trends in Cultivators v/s Agricultural Workers",fontsize=14,weight="bold")
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

# Taking Insights into Relegious Based Population

```python
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/Census_2011")
querry="""
SELECT State_name,
       SUM(Hindus) / SUM(Population) * 100 AS Hindu_Percentage,
       SUM(Muslims) / SUM(Population) * 100 AS Muslim_Percentage,
       SUM(Christians) / SUM(Population) * 100 AS Christian_Percentage,
       SUM(Sikhs) / SUM(Population) * 100 AS Sikh_Percentage,
       SUM(Buddhists) / SUM(Population) * 100 AS Buddhist_Percentage,
       SUM(Jains) / SUM(Population) * 100 AS Jain_Percentage
FROM census
GROUP BY State_name
ORDER BY State_name;
"""
df=pd.read_sql(querry,engine)
engine.dispose()

df_melted = df.melt(id_vars=["State_name"], 
                    value_vars=["Hindu_Percentage", "Muslim_Percentage", "Christian_Percentage", 
                                "Sikh_Percentage", "Buddhist_Percentage", "Jain_Percentage"], 
                    var_name="Religion", value_name="Religion_Percentage")

plt.figure(figsize=(25,12))
sns.barplot(data=df_melted,x="State_name", y="Religion_Percentage",hue="Religion")
plt.xlabel("State name",fontsize=12,weight="bold")
plt.ylabel("Religios Percentage",fontsize=12,weight="bold")
plt.title(" Taking Insights into Relegious Based Population",fontsize=14,weight="bold")
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
