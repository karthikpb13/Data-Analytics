# Q8) -- Calculate total no. of houses sold per area

```python
from sqlalchemy import create_engine
import pandas as pd

# Create an SQLAlchemy engine for MySQL
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/london_housing")

query = """
SELECT area, SUM(houses_sold) AS total_houses_sold
FROM house_data
GROUP BY area;
"""
df = pd.read_sql(query, engine)

# Close the engine after use (optional in most cases)
engine.dispose()
# Plotting the data (Bar Plot)
plt.figure(figsize=(12, 6))
sns.barplot(x='area', y='total_houses_sold', data=df)
plt.title('Total Houses Sold by Area')
plt.xlabel('Area')
plt.ylabel('Total Houses Sold')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

# Q9) Find the average house price by area 

```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create an SQLAlchemy engine for MySQL
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/london_housing")

# Corrected query string
query = """
select area, avg(average_price) as avg_house_price
from house_data
group by area;
"""

df = pd.read_sql(query, engine)

# Close the engine after use (optional in most cases)
engine.dispose()

# Plotting the data
plt.figure(figsize=(14, 7))
sns.set(style="whitegrid")

# Bar plot showing average house price by area
sns.barplot(x='area', y='avg_house_price', data=df, palette="coolwarm", hue='area')
plt.title("Average house price by area", fontsize=14, weight="bold")
plt.xlabel("Area")
plt.ylabel("Average House Price")
plt.xticks(rotation=45, ha='right', fontsize=8)

# Adding data labels on top of bars for clarity
for index, row in df.iterrows():
    plt.text(row.name, row.avg_house_price, round(row.avg_house_price, 2), color='black', ha="center", va="bottom", fontsize=12)

plt.grid(True, axis='y', linestyle='--', alpha=0.7)  # Adding grid for better readability

plt.tight_layout()  # Adjust the layout to prevent overlap
plt.show()
```

## Q10) Find Areas with the Highest Crime Rates

```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create an SQLAlchemy engine for MySQL
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/london_housing")

# SQL query to fetch the data
query = """
SELECT area, year, AVG(average_price) AS avg_house_price
FROM house_data
GROUP BY year, area
ORDER BY year;
"""

# Read data into a pandas DataFrame
df = pd.read_sql(query, engine)

# Close the engine after use
engine.dispose()

# Plotting the data
plt.figure(figsize=(30,30))
sns.set(style="whitegrid")

# Line plot with 'year' on the x-axis, 'avg_house_price' on the y-axis, and hue for 'area'
sns.lineplot(data=df, x='year', y='avg_house_price', hue='area', marker="o", palette="tab10")

# Adding plot details
plt.title("Average House Price by Area Over Years", fontsize=16, weight="bold")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average House Price", fontsize=12)
plt.legend(title="Area", loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

# Adjust layout to fit everything nicely
plt.tight_layout()

# Show the plot
plt.show()
```

## Q11)
### What are the yearly trends in average house prices across different areas of London?
### Which area has shown the highest increase in average house prices over the last decade?
### What are the areas with the highest increase in the average price 

```python
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/london_housing")

query_trends = """
SELECT year, area, AVG(average_price) AS avg_house_price
FROM house_data
GROUP BY year, area
ORDER BY year, area;
"""

# Query for area with highest increase in average house prices
query_highest_increase = """
WITH price_trends AS (
    SELECT area, year, AVG(average_price) AS avg_house_price
    FROM house_data
    GROUP BY year, area
)
SELECT area, MAX(avg_house_price) - MIN(avg_house_price) AS price_increase
FROM price_trends
GROUP BY area
ORDER BY price_increase DESC
LIMIT 10;
"""

# Execute the queries
df_trends = pd.read_sql(query_trends, engine)
df_highest_increase = pd.read_sql(query_highest_increase, engine)
engine.dispose()

# plotting yearly trend
plt.figure(figsize=(10,8))
sns.lineplot(data=df_trends,x='year',y='avg_house_price', hue='area', marker='o')
plt.title("Yearly Trends in Average House price", fontsize=14,weight="bold")
plt.xlabel("Year",fontsize=10)
plt.ylabel("Average house Price",fontsize=10)
plt.xticks(rotation=90)
plt.legend(loc="upper left", title="Area", bbox_to_anchor=(1,1))
plt.tight_layout()
plt.show()

print("Area with the highest increase in average house prices:", df_highest_increase["area"][0])
```

### Area with highest increase in House Price

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import pandas as pd

# SQL query for the highest increase in average house prices
query_highest_increase = """
WITH price_trends AS (
    SELECT area, year, AVG(average_price) AS avg_house_price
    FROM house_data
    GROUP BY year, area
)
SELECT area, MAX(avg_house_price) - MIN(avg_house_price) AS price_increase
FROM price_trends
GROUP BY area
ORDER BY price_increase DESC
LIMIT 10;
"""

# Create SQLAlchemy engine and execute query
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/london_housing")
df_highest_increase = pd.read_sql(query_highest_increase, engine)
engine.dispose()

# Plotting the highest increase in house prices
plt.figure(figsize=(12, 8))
sns.barplot(
    x="price_increase", 
    y="area", 
    data=df_highest_increase, 
    palette="viridis",
    hue="price_increase"
)

plt.title("Top 10 Areas with Highest Increase in House Prices", fontsize=16, weight="bold")
plt.xlabel("Price Increase", fontsize=12)
plt.ylabel("Area", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

# Q12) 
## How does the crime rate in an area correlate with house prices?
## Do areas with higher crime rates see a slower increase in house prices compared to areas with lower crime rates?
## Is there a relationship between the number of crimes and the number of houses sold in a specific area?

```python

engine = create_engine("mysql+pymysql://root:1234@localhost:3306/london_housing")
# Load data for correlation analysis
query_crime_prices = """
WITH crime_house_data AS (
    SELECT 
        area, 
        AVG(average_price) AS avg_house_price, 
        SUM(no_of_crimes) AS total_crimes
    FROM house_data
    GROUP BY area
)
SELECT 
    area, 
    avg_house_price, 
    total_crimes
FROM crime_house_data;
"""

df_crime_prices = pd.read_sql(query_crime_prices, engine)
engine.dispose()

# Correlation Analysis
correlation = df_crime_prices['total_crimes'].corr(df_crime_prices['avg_house_price'])
print(f"Correlation between crime rate and house prices: {correlation:.2f}")

# Scatter plot for crime vs house prices
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_crime_prices, x='total_crimes', y='avg_house_price', hue='area', palette='viridis', s=100)
plt.title("Correlation Between Crime Rates and House Prices", fontsize=14, weight="bold")
plt.xlabel("Total Crimes", fontsize=12)
plt.ylabel("Average House Price", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
